```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import smtplib, ssl
import schedule


    

def job():
    
    port = 465 
    
    #insert password yourself
    
    password = "LOREM IPSUM "
    smtp_server = "smtp.gmail.com"

    #message and send, recieve email addresses -- insert yourself
    sender = "LOREM IPSUM"
    receiver = "LOREM IPSUM"
    message1 = "found a Corgi!!!!"
    message2 = "found a Shih Tzu!!!!"
    message3 = "found a Maltese!"

    # Create a secure SSL context, initiate smtp connection
    server = smtplib.SMTP_SSL("smtp.gmail.com", port)
    server.ehlo()
    server.login(sender, password)



    #initiate driver and define wait between actions
    link = "http://petharbor.com/results.asp?searchtype=ADOPT&start=3&friends=0&samaritans=0&nosuccess=0&rows=10&imght=120&imgres=thumb&tWidth=200&view=sysadm.v_miad&text=000000&fontface=arial&fontsize=10&col_bg=99b5c9&col_bg2=e7eec4&SBG=026BA9&zip=33183&miles=10&shelterlist=%27MIAD%27&atype=&where=type_DOG,gender_m,size_s&PAGE=1"
    driver = webdriver.Chrome()
    driver.get(link)
    wait = WebDriverWait(driver, 1000)


    #click to relevant page
    wait
    driver.find_element_by_xpath('//*[@id="frmResults"]/table[3]/tbody/tr[1]/td[4]/a').click()
    wait

    #navigate to relevant page

    potentialClicks = driver.find_element_by_xpath('//*[@id="frmResults"]/center[3]').find_elements_by_tag_name("a")
    size = len(potentialClicks)
    x = size - 1
    driver.find_element_by_xpath( '//*[@id="frmResults"]/center[3]/a[%s]' %x).click()
    

    #find text elements with dog descriptions
    wait
    possible_list = driver.find_elements_by_tag_name('td')

    #iterate through options to find desired breed and send email if found
    for element in possible_list:
        if element.text.__contains__("Shih Tzu"):
            server.sendmail(sender, receiver, message1)
        elif element.text.__contains__("Corgi"):
            server.sendmail(sender, receiver, message2)
        elif element.text.__contains__("Maltese"):
            server.sendmail(sender, receiver, message3)

    server.close()


#scheduler to automate runs
schedule.every().day.at("16:00").do(job)
schedule.every().day.at("18:00").do(job)
schedule.every().day.at("20:00").do(job)

while True:
    schedule.run_pending()
    time.sleep(60) # wait one minute
        
```
